```python
#NLP Dataset vectorization and modelling with 6 classifiers

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# Load data from the CSV file
data_new = pd.read_csv(r"C:\Users\Tombra\New folder\data_new.csv")

# Extract the combined text column to be vectorized
text_data = data_new["DESCRIPTION"]

# Create the TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Vectorize the text data
X = vectorizer.fit_transform(text_data)

# Get the class labels
labels = data_new["ROCK NAME"]

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

# Train a Random Forest classifier
rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train, y_train)

# Train a Logistic Regression classifier
lr_classifier = LogisticRegression()
lr_classifier.fit(X_train, y_train)

# Train a Gaussian Naive Bayes classifier
nb_classifier = GaussianNB()
nb_classifier.fit(X_train.toarray(), y_train)

# Train a Gradient Boosting classifier
gb_classifier = GradientBoostingClassifier()
gb_classifier.fit(X_train, y_train)

# Train a K-Nearest Neighbors classifier
knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train, y_train)

# Train a Support Vector Classifier
svc_classifier = SVC(probability=True)
svc_classifier.fit(X_train, y_train)

# Predict the labels for the test set using Random Forest
y_pred_rf = rf_classifier.predict(X_test)

# Predict the labels for the test set using Logistic Regression
y_pred_lr = lr_classifier.predict(X_test)

# Predict the labels for the test set using Gaussian Naive Bayes
y_pred_nb = nb_classifier.predict(X_test.toarray())

# Predict the labels for the test set using Gradient Boosting
y_pred_gb = gb_classifier.predict(X_test)

# Predict the labels for the test set using K-Nearest Neighbors
y_pred_knn = knn_classifier.predict(X_test)

# Predict the labels for the test set using Support Vector Classifier
y_pred_svc = svc_classifier.predict(X_test)

# Calculate accuracy for each classifier
accuracy_rf = accuracy_score(y_test, y_pred_rf)
accuracy_lr = accuracy_score(y_test, y_pred_lr)
accuracy_nb = accuracy_score(y_test, y_pred_nb)
accuracy_gb = accuracy_score(y_test, y_pred_gb)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
accuracy_svc = accuracy_score(y_test, y_pred_svc)

# Display the accuracy for each classifier
print("Random Forest Classifier Accuracy:", accuracy_rf)
print("Logistic Regression Classifier Accuracy:", accuracy_lr)
print("Naive Bayes Classifier Accuracy:", accuracy_nb)
print("Gradient Boosting Classifier Accuracy:", accuracy_gb)
print("K-Nearest Neighbors Classifier Accuracy:", accuracy_knn)
print("Support Vector Classifier Accuracy:", accuracy_svc)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# Load the combined results
combined_results_path = r"C:\Users\Tombra\New folder\combined_results_new_4a.csv"
combined_results = pd.read_csv(combined_results_path)

# Split the data into features (probabilities) and target (ROCK NAME)
features = combined_results.iloc[:, 1:]  # Exclude the ROCK NAME column
target = combined_results["ROCK NAME"]

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Train the individual classifiers
rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train, y_train)

svc_classifier = SVC()
svc_classifier.fit(X_train, y_train)

knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train, y_train)

# Create the ensemble model with the trained classifiers
ensemble_model = VotingClassifier(
    estimators=[
        ('random_forest', rf_classifier),
        ('support_vector', svc_classifier),
        ('k_nearest_neighbors', knn_classifier)
    ],
    voting='hard'  # Use majority voting
)

# Train the ensemble model
ensemble_model.fit(X_train, y_train)

# Evaluate the ensemble model
ensemble_accuracy = ensemble_model.score(X_test, y_test)
print("Ensemble Model Accuracy:", ensemble_accuracy)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# Load the combined results
combined_results_path = r"C:\Users\Tombra\New folder\class_probabilities.csv" #raw cnn probabilities
combined_results = pd.read_csv(combined_results_path)


# Split the data into features (probabilities) and target (ROCK NAME)
features = combined_results.iloc[:, 1:]  # Exclude the ROCK NAME column
target = combined_results["ROCK NAME"]

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Train the individual classifiers
rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train, y_train)

svc_classifier = SVC()
svc_classifier.fit(X_train, y_train)

knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train, y_train)

# Create the ensemble model with the trained classifiers
ensemble_model = VotingClassifier(
    estimators=[
        ('random_forest', rf_classifier),
        ('support_vector', svc_classifier),
        ('k_nearest_neighbors', knn_classifier)
    ],
    voting='hard'  # Use majority voting
)

# Train the ensemble model
ensemble_model.fit(X_train, y_train)

# Evaluate the ensemble model
ensemble_accuracy = ensemble_model.score(X_test, y_test)
print("Ensemble Model Accuracy:", ensemble_accuracy)
```
