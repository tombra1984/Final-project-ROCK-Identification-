```python
#combining the old CNN probabilites out put with the new NLP probabilities 53_CLASSES_DESCRIPTION_3 to get one output using averaged probabilities



import pandas as pd

# Load the probability results from the RF classifier
rf_results_path = r"C:\Users\Tombra\New folder\results_new_4.csv"
rf_results = pd.read_csv(rf_results_path)

# Load the probability results from the EfficientNet model
cnn_results_path = r"C:\Users\Tombra\New folder\COMBINE_CNN.csv"
cnn_results = pd.read_csv(cnn_results_path)

# Merge the results based on the class names
merged_results = pd.merge(rf_results, cnn_results, on="ROCK NAME")

# Get the list of class names
class_names = merged_results["ROCK NAME"]

# Calculate the average probabilities for each class
avg_probabilities = merged_results.iloc[:, 1:].mean(axis=1)

# Create a new DataFrame with the combined results
combined_results = pd.DataFrame({
    "ROCK NAME": class_names,
    "AVG PROBABILITY": avg_probabilities
})

# Sort the results by average probability in descending order
combined_results = combined_results.sort_values(by="AVG PROBABILITY", ascending=False)

# Display the combined results
print("Combined Results:")
print(combined_results)

# Save the combined results to a CSV file
output_path = r"C:\Users\Tombra\New folder\combined_results_new_4a.csv"
combined_results.to_csv(output_path, index=False)

print("Combined results saved successfully.")
```
